```python
import pandas as pd
import numpy as np

exam_data = {
    'name': ['Anastasia', 'Dima', 'Kathe', 'James', 'Emily',
             'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no',
                'yes', 'yes', 'no', 'no', 'no']
}

labels = ['a','b','c','d','e','f','g','h','i','j']

df = pd.DataFrame(exam_data, index=labels)

print(df[['name', 'score']])
```

```python
print(df[df['score'].isna()])
```

```python
# Append new row
df.loc['k'] = ['Suresh', 15.5, 1, 'yes']
print("After appending new row:\n")
print(df)

# Delete the new row
df = df.drop('k')
print("\nAfter deleting new row:\n")
print(df)
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\KIIT0001\Downloads\student_marks_10.xlsx")

df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
df["Average"] = df["Total"] / 4

df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print(df)
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\KIIT0001\Downloads\employee_salary_10.xlsx")

df["New_Salary"] = df.apply(
    lambda row: row["Salary"] * 1.10 if row["Experience"] >= 5 else row["Salary"] * 1.07,
    axis=1
)

print(df)
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\KIIT0001\Downloads\sales_data_10.xlsx")

result = df.groupby("Product")["Sales"].agg(
    Total_Sales="sum",
    Average_Sales="mean"
)

print(result)
```
