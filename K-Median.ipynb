```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# create random dataset
X = np.random.rand(100,2)

plt.scatter(X[:,0], X[:,1])
plt.title("Original Data")
plt.show()
```

```python
k = 3   # number of clusters

# randomly choose initial medians
medians = X[np.random.choice(len(X), k, replace=False)]

for iteration in range(100):

    # calculate Manhattan distance
    distances = np.sum(np.abs(X[:, np.newaxis] - medians), axis=2)

    # assign cluster
    labels = np.argmin(distances, axis=1)

    # update medians
    new_medians = np.array([np.median(X[labels == i], axis=0) for i in range(k)])

    # stop if medians do not change
    if np.all(medians == new_medians):
        break

    medians = new_medians
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow')
plt.scatter(medians[:,0], medians[:,1], c='black', marker='X', s=200)
plt.title("K-Median Clustering")
plt.show()
```
